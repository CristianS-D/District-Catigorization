```python
page = bs.BeautifulSoup(url_link.text, 'html.parser')
tables = page.find_all('table')
len(tables)
tables[0]
```

```python
key_table = tables[0]
```

## Extracting

```python
rows = key_table.find_all('tr')
len(rows)
```

```python
rows[0].find_all(['th', 'td'])
```

```python
cells = rows[0].find_all(['th', 'td'])
[cell.get_text(strip=True) for cell in cells]
```

```python
all_rows = []
for row in rows:
    cells = row.find_all(['th', 'td'])
    all_rows.append([cell.get_text(strip=True) for cell in cells])
```

```python
all_rows
```

```python
df = pd.DataFrame(all_rows)
df
```

```python
df.columns = df.iloc[0]   # set headers
#df = df.iloc[1:]          # drop header row
#df = df.reset_index(drop=True)
df
```

```python
df = df.iloc[1:]
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
title = page.find_all('title')
title_text = title[0].get_text(strip=True)
title_text
```

```python
safe_title = re.sub(r'[^\w\s-]', '', title_text)  
safe_title = safe_title.strip().replace(' ', '_')
safe_title
```

```python
filename = f"{safe_title}_Staff.csv"
df.to_csv(f'Staff & Student Keys\{filename}', index=False)
```
